## Data Visualization Modulus - Python Format

### Import relevant packages: 
#### *matplotlib.pyplot* is for plotting 
#### *numpy* is for easier data manipulation 
#### *pandas* is for loading in .csv files

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# We can immediately plot this data now using the python library matplotlib
# that we imported in the first cell. To do this, we first need to create an 
# empty figure, which has empty axes. We will build this figure up step by 
# step. 

# To initialize a figure, we use the following one line of code. This creates a
# figure with no data on it and a set of axes that are unlabeled.
figure, axes = plt.subplots()
```

### Now that we have a figure and some axes, we can start to build this figure up to include labels and data. We know that our data includes ice coverage with time, so let's get our figure looking better by adding some labels to the axes and a title. 

```python
# First, recreate a figure and axes
figure, axes = plt.subplots()

# Now, we want to label the x and y axes. On the x-axis, we will plot time, and on the 
# y-axis, we will plot ice coverage.

# Set the label on the x-axis
axes.set_xlabel("Date")

# Set the label on the y-axis
axes.set_ylabel("Ice Coverage")

# Set the title
axes.set_title("Ice Coverage vs Time")
```

#### Now that our figure has labeled axes and a title, we need to load in the data so that we can plot it.

```python
# Specify the filename
filename = "data.csv"

# Load in the data. The first 4 rows are text which describes
# the dataset, so we skip these 4 rows
raw_data = pd.read_csv(filename, delimiter=',', skiprows=4)

# Display what different columns are within the loaded data
print(raw_data.keys())
```

#### For the first plot, we want the time, which is stored in 'Date', and the ice coverage, which is stored in 'Value'.

```python
# Extract the date column and the value columns for plotting:
Dates  = raw_data["Date"].to_numpy()
Values = raw_data["Value"].to_numpy()

# Create the same figure that we created above:

figure, axes = plt.subplots()

axes.set_xlabel("Date")
axes.set_ylabel("Ice Coverage")
axes.set_title("Ice Coverage vs Time")

# Now, add the data to the plot. Matplotlib will automatically set the 
# values on the x-axis and the y-axis to match the dataset. We can do a 
# simple scatter plot with the following one line of code:
axes.scatter(Dates, Values)
```

#### We have data on the plot! But you'll quickly notice that the data doesn't look right. The dates are nonsensical, and there are two values that have an area of -10,000. The negative values were put into the dataset as placeholders for times when there were *NO* data, so we need to remove these points.

```python
# Now, extract the date column and the value columns for plotting:
Dates  = raw_data["Date"].to_numpy()
Values = raw_data["Value"].to_numpy()

# Within the values dataset, there are certain values that are filled in
# for areas where there is no data. These are values initialized with a 
# sea ice coverage of -9999. Remove these entries from the data before 
# plotting. 

valid_indices = Values >= 0

figure, axes = plt.subplots()

axes.set_xlabel("Date")
axes.set_ylabel("Ice Coverage")
axes.set_title("Ice Coverage vs Time")

# Now, add the data to the plot. Matplotlib will automatically set the 
# values on the x-axis and the y-axis to match the dataset. We can do a 
# simple scatter plot with the following one line of code:
axes.scatter(Dates[valid_indices] / 100, Values[valid_indices])

plt.show()
```

#### This looks much better. With just the raw data, is there anything you can say about the ice coverage with time?

#### A scatter plot isn't the only way to present data in a figure. Common other ways to present data is with a curve, or with a histogram (bar plot). Let's use this same dataset but change the way we plot it. First, we'll plot the data as a curve. 

```python
figure, axes = plt.subplots()

axes.set_xlabel("Date")
axes.set_ylabel("Ice Coverage")
axes.set_title("Ice Coverage vs Time")

# We can simply change the way that the data is plotted from a scatter plot to a curve
# by changing the function from 'scatter' to 'plot'.
axes.plot(Dates[valid_indices] / 100, Values[valid_indices])

plt.show()
```

# Now, let's plot it as a histogram.

```python
figure, axes = plt.subplots()

# A histogram does not display time on the x-axes, so we need to change the axes labels
# for our plot.
axes.set_xlabel("Ice Coverage")
axes.set_ylabel("Count")
axes.set_title("Ice Coverage Histogram")

# We can simply change the way that the data is plotted from a scatter plot to a curve
# by changing the function from 'scatter' to 'plot'.
axes.hist(Values[valid_indices])
plt.show()
```
